```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from main_folder.get_data import get_data

```

```python
data = get_data()
print(type(data), '\n', 'Index:', data.index, '\n', 'Columns:',data.columns)
```

## The Data
Let's investugate the data by using the Pandas library

```python
data.head(5)
```

We can see that there are several models off the samebrand, at the same
time just having the model name might not be suficient. Therfore we
create a new column combining them both.

```python
data['Brand_Model'] = data['Brand'] + data['Model']
data[['Brand_Model']].head(3)
```

Sorting intresting attributes

```python
range_data = data.sort_values(by=['Range_Km'], ascending=False)
effi_data = data.sort_values(by=['Efficiency_WhKm'], ascending=False)
topspeed_data = data.sort_values(by=['TopSpeed_KmH'], ascending=False)


```

### Range plot

```python
PLOT_RANGE = 10

fig, ax = plt.subplots()
ax1 = sns.barplot(y=range_data['Range_Km'].head(PLOT_RANGE), 
                  x=range_data['Brand_Model'].head(PLOT_RANGE))
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
                 #hue=data['Brand'].head(PLOT_RANGE))
#ax.set_ylim(ax.get_ylim())
#ax.set_ylabel('Range [km]')
```

### Energy Efficiency

```python
PLOT_RANGE = 10


fig, ax = plt.subplots()
ax2 = sns.barplot(y=effi_data['Efficiency_WhKm'].tail(PLOT_RANGE), 
                  x=effi_data['Brand_Model'].tail(PLOT_RANGE).values)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
plt.plot()
                  
```

```python

fig, ax 
sns.set_color_codes("pastel")
fig = sns.barplot(x='Range_Km', y='Brand_Model', data = range_data,
            label="Range [km]", color="b")

sns.set_color_codes("muted")
sns.barplot(x='Efficiency_WhKm', y='Brand_Model', data = range_data,
            label='Efficiency WhKm', color="b")

plt.rcParams['figure.figsize'] = [6, 30]
```

```python
PLOT_RANGE = 6
data_plot= range_data
fig, ax = plt.subplots()
ax = sns.scatterplot(x=data_plot['Range_Km'].head(PLOT_RANGE), 
                     y=data_plot['Efficiency_WhKm'].head(PLOT_RANGE), 
                     hue=data_plot['Brand_Model'].head(PLOT_RANGE))
ax.set_ylim(ax.get_ylim()[::-1])
ax.set_xlabel('Range [km]')
ax.set_ylabel('Efficiency [Wh/km]')
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot()
```
