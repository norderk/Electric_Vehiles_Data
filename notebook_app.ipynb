```python
# Sorting the data on interesting attributes
range_data = data.sort_values(by=['Range_Km'], ascending=False)
effi_data = data.sort_values(by=['Efficiency_WhKm'], ascending=False)
topspeed_data = data.sort_values(by=['TopSpeed_KmH'], ascending=False)
price_data = data.sort_values(by=['PriceEuro'],ascending=False)


```

### Range plot

```python
#Visualizing the top 10 the vehicle types based on range 
PLOT_RANGE = 10

fig, ax = plt.subplots()
ax1 = sns.barplot(y=range_data['Range_Km'].head(PLOT_RANGE), 
                  x=range_data['Brand_Model'].head(PLOT_RANGE))
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
                 #hue=data['Brand'].head(PLOT_RANGE))
#ax.set_ylim(ax.get_ylim())
#ax.set_ylabel('Range [km]')
```

### Energy Efficiency

```python
#Visualizing the top 10 the vehicle types based on efficiency

PLOT_RANGE = 10


fig, ax = plt.subplots()
ax2 = sns.barplot(y=effi_data['Efficiency_WhKm'].tail(PLOT_RANGE), 
                  x=effi_data['Brand_Model'].tail(PLOT_RANGE).values)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
plt.plot()
                  
```
