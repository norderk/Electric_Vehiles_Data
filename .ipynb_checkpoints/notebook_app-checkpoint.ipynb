```python
data['Brand_Model'] = data['Brand'] + data['Model']
data[['Brand_Model']].head(3)
```

*Table2:* Creating a new columne with a combine name of Brand and Model, to get a more divers sorting option

```python
# Sorting the data on interesting attributes
range_data = data.sort_values(by=['Range_Km'], ascending=False)
effi_data = data.sort_values(by=['Efficiency_WhKm'], ascending=False)
topspeed_data = data.sort_values(by=['TopSpeed_KmH'], ascending=False)
price_data = data.sort_values(by=['PriceEuro'],ascending=False)


```

### Range plot

```python
#Visualizing the top 10 the vehicle types based on range 
PLOT_RANGE = 10

fig, ax = plt.subplots()
ax1 = sns.barplot(y=range_data['Range_Km'].head(PLOT_RANGE), 
                  x=range_data['Brand_Model'].head(PLOT_RANGE))
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
                 #hue=data['Brand'].head(PLOT_RANGE))
#ax.set_ylim(ax.get_ylim())
#ax.set_ylabel('Range [km]')
```

### Energy Efficiency

```python
#Visualizing the top 10 the vehicle types based on efficiency

PLOT_RANGE = 10


fig, ax = plt.subplots()
ax2 = sns.barplot(y=effi_data['Efficiency_WhKm'].tail(PLOT_RANGE), 
                  x=effi_data['Brand_Model'].tail(PLOT_RANGE).values)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = [8, 4]
plt.gca().invert_xaxis()
plt.plot()
                  
```

*Figure:* Light-Blue is Range [km], Dark-blue is the Energy Efficiency [Wh/km]

```python
#Visualizing the top 10 the vehicle types based on efficiency
PLOT_RANGE = 10


fig, ax = plt.subplots()
ax2 = sns.barplot(y=price_data['PriceEuro'].tail(PLOT_RANGE), 
                  x=price_data['Brand_Model'].tail(PLOT_RANGE).values)
plt.xticks(rotation=90)
plt.gca().invert_xaxis()
plt.rcParams['figure.figsize'] = [8, 4]
plt.plot()
```

### Putting it all toghether

```python
PLOT_RANGE = 10
data_plot= range_data
fig, ax = plt.subplots()

colors = np.array(data_plot['PriceEuro'].head(PLOT_RANGE).values)

lst = data_plot['Brand_Model'].head(PLOT_RANGE).values


ax = plt.scatter(x=data_plot['Range_Km'].head(PLOT_RANGE), 
                     y=data_plot['Efficiency_WhKm'].head(PLOT_RANGE),
                     s=200,
                     c=colors,
                     cmap='coolwarm')
for i in range(0, len(data_plot.head(PLOT_RANGE))):
    plt.annotate(lst[i], (data_plot['Range_Km'].iloc[i]+10, data_plot['Efficiency_WhKm'].iloc[i] + 5))


plt.gca().invert_yaxis()
plt.xlabel('Range [km]')
plt.ylabel('Energy Efficiency [Wh/km]')
plt.colorbar(label='Price [EURO]')
plt.plot()


```

*Figure:* Based of using range as the primary dimension and efficiency and price as secondary, these are the top 10 choices


```python
PLOT_RANGE = 10
data_plot= effi_data.tail(PLOT_RANGE).sort_values(by='Efficiency_WhKm', ascending=True)
fig, ax = plt.subplots()

colors = np.array(data_plot['PriceEuro'].values)

lst = data_plot['Brand_Model'].values
lst_pos = [2,0,2,-2,-2,2,-2,-2,2,1]


ax = plt.scatter(x=data_plot['Range_Km'], 
                     y=data_plot['Efficiency_WhKm'],
                     s=100,
                     c=colors,
                     cmap='coolwarm')

for i in range(0, len(data_plot)):
    # print(lst[i])
    plt.annotate(lst[i], (data_plot['Range_Km'].iloc[i], data_plot['Efficiency_WhKm'].iloc[i]-lst_pos[i]))

plt.gca().invert_yaxis()
plt.xlabel('Range [km]')
plt.ylabel('Energy Efficiency [Wh/km]')
plt.colorbar(label='Price [EURO]')
plt.rcParams['figure.figsize'] = [16, 8]
plt.plot()
```

*Figure:* Based of using efficiency as the primary dimension and range and price as secondary, these are the top 10 choices
